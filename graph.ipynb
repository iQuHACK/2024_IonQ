```python
import rustworkx 
import matplotlib.pyplot as plt
from rustworkx.visualization import mpl_draw
```

```python
graph = rustworkx.PyGraph()

graph.add_node("q0")
graph.add_node("q1")
graph.add_node("q2")

potential_node_dic = {0 : [1,2], 1 : [3,4], 2: [5,6]}
```

```python
def visualize_graph(graph):
    # Create a larger figure with a specified size (adjust the values as needed)
    # fig = plt.figure(figsize=(40, 40))

    # Create a subplot within the larger figure
    subax1 = plt.subplot(121)

    # Now, you can draw your graph with_labels=True on the larger subplot
    mpl_draw(graph, with_labels=True, ax=subax1)
```

```python
def connect(q0, qx, graph, potential_node_dic):
    graph.add_edge(q0, qx, 1)
    
    if qx in potential_node_dic.keys():   
        for node in potential_node_dic[qx]:
            potential_node_dic[q0].append(node)
            graph.add_node(node)

    potential_node_dic[q0].remove(qx)
```

```python
visualize_graph(graph)
```

```python
connect(0, 1, graph ,potential_node_dic)
```

```python
visualize_graph(graph)
```

```python
connect(0, 2, graph ,potential_node_dic)
```

```python
visualize_graph(graph)
```
